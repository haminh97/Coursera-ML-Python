#### Visualizing Data

```python
data = genfromtxt('ex1data1.txt', delimiter=',')

X = data[:, 0]
Y = data[:, 1]
m = X.shape[0]
X = np.reshape(X, (m, 1))
Y = np.reshape(Y, (m, 1))

print('First 10 data points\n')
print(data[0:10,:])
```

In the first part of Excercise 1, we are given the problem of fitting a linear regression to a set of data. The first column (our X value) is the population of a city in 10000s and the second column (our Y value) is the profit of a food truck in that city in 1000$. A negative value for profit simply indicates a loss. With those information in mind, let's plot our data on a simple 2D plane, with the X-axis be the cities' population, and the Y-axis be the corresponding profits. We do this using matplotlib:

```python
# Plot the data
plt.figure()
plt.plot(X, Y, 'rx', ms=10)
plt.xlabel('Population of city in 10000s')
plt.ylabel('Profit in $10000')
plt.show()
```

The visualization shows that there is a positive correlation between the cities' population and the profits. How should we use that information to predict new profits in a new city not in this dataset?

#### Linear Regression-the basic

The solution is simple: We can fit a straight line through our dataset. This line should be the best line that we can use to predict our profit given the size of a city. In other word, if I were given the size of the city, I should be able to predict the profit a food truck within that city as accurate as possible. This is called an **hypothesis**, otherwise written as $h_\theta(x)$ (shorthand: $h_x$)

Using only one variable, our hypothesis should then be $h_\theta(x) = \theta_0 \cdot x_0 + \theta_1 \cdot x_1$, or, if you prefer matrix multiplication, $\begin{bmatrix} \theta_0 & \theta_1 \end{bmatrix} \cdot \begin{bmatrix}  x_0 \\ x_1  \end{bmatrix} $ (remember that $x_0 = 1$)

Let's fit an arbitray hypothesis over this dataset and see what's happening:

```python
plt.figure()
plt.plot(X, Y, 'rx', ms=10)
plt.plot(X, np.dot(np.column_stack((np.ones(m), X)), np.array([[1],[3]])), '-')
plt.xlabel('Population of city in 10000s')
plt.ylabel('Profit in $10000')
plt.legend(['Training data', 'Our hypothesis'])
plt.show()
```

Not quite predictive of our model, isn't it? So how can we find out the line that can best fit this dataset? First of all, we need a measurement of how "off" a hypothesis is. 

#### Cost Function

Cost Function is literary what it is--a function that compute the cost, e.g. error, between our hypothesis and our actual data. One could think of the simplest cost function as 

\begin{equation} COST = \text{predicted value - actual value} \end{equation} 

and we can sum them up over all of our training data to find the total cost of the hypothesis.

$ \sum\limits^{m}_{i = 1} (hypothesis^{(i)} - actual^{(i)}) $

However, there is one problem, if we sum our error like that over all of our training data, some negative and positive data will cancel out. To counter this, there are two ways we can sum our errors:

1. Summing the absolute values:
$ \sum\limits^{m}_{i = 1} \left| hypothesis^{(i)} - actual^{(i)} \right| $
2. Summing the square differences:
$ \sum\limits^{m}_{i = 1} (hypothesis^{(i)} - actual^{(i)})^2 $

If you are good with calculus and statistic, you will see that the second way, summing the square differences, will have some very nice properties with regard to derivation. To understand more, you can visit this [StackExchange question](https://math.stackexchange.com/questions/967883/why-get-the-sum-of-squares-instead-of-the-sum-of-absolute-values) on why we prefer to square the sum instead of taking the absolute values.

To get the actual cost function, we divide the square differences by 2m. Such an arbitrary number is choosen because when we derive this cost function, the 2 will cancel out. We will see this in a moment.

Finally, we have our cost function as follows. Remember that $h_\theta(x^{(i)})$ and $y^{(i)}$ refers to the i-th predicted hypothsis and i-th actual value in the training set:

\begin{equation} J = \frac{1}{2m} \cdot \sum\limits^{m}_{i = 1} (h_\theta(x^{(i)}) -  y^{(i)})^2 \end{equation}

```python
def compute_cost(X, Y, theta):
    m = X.shape[0]
    return 1 / (2 * m) * np.sum(np.square(np.dot(X, theta) - Y))
```

```python
X = np.column_stack((np.ones(m), X))
theta = np.zeros((2, 1))

print('With theta = [0 ; 0]\nCost computed = %f\n' % compute_cost(X, Y, theta))

print('\nWith theta = [-1 ; 2]\nCost computed = %f\n' % compute_cost(X, Y, np.array([[-1], [2]])))
```

That's great for us to know how "off" a hypothesis is from the actual data, but how can we use this to fit our best line?

#### Gradient Descent

Certainly, we should make our cost as low as possible. If you learn calculus, you'd probably know a way to find the minimum value could be taking the derivative and setting it to zero. And we can quite do so with this cost function. However, when the cost function get really complicated, how can we optimize our cost function to a minimum value? There exist a field of mathematics dedicated to this problem call [Mathematical Optimization](https://en.wikipedia.org/wiki/Mathematical_optimization). A popular optimization algorithm is called [Gradient Descent](https://en.wikipedia.org/wiki/Gradient_descent), which we will implement in the code below.

Put simply, gradient descent is taking the first order derivative repeatedly and using that derivative to determine where the value should go. A good analogy is, say, imagining a blind man trying to navigate a hill. If he want to find the top of the hill (local maxima), he will have to walk toward where the hill goes up (follow the direction with positive derivation) at a certain walking speed (our gradient descent speed). Vice versa, if he want to find the foot of the hill (local minima), he will have to walk toward where the hill go down (follow the direction with negative derivation) at a certain walking speed (gradient descent speed). An algorithm of the gradient descent function would then be:

$\quad\text{Repeat: } \theta_j = \theta_j - \alpha \frac{\partial J}{\partial \theta_j} \quad \text{for every j}$

In here, we see that $\frac{\partial J}{\partial \theta_j}$ is the partial derivative of the cost function with respect to $\theta_j$, and $\alpha$ is our gradient descent 'speed', or our learning rate. 

Now, we should implement the gradient descent algorithm to minimize our cost function. To do so, first we need to find the partial derivative of the cost function $\frac{\partial J}{\partial \theta_j}$

\begin{align*}
\frac{\partial}{\partial \theta_j} 
&= \frac{\partial}{\partial \theta_j} \Big(\frac{1}{2m} \sum\limits^{m}_{i = 1} (h_\theta(x^{(i)}) -  y^{(i)})^2\Big) \\
&= \frac{1}{2m} \sum\limits^{m}_{i = 1} \frac{\partial}{\partial \theta_j} \Big(h_\theta(x^{(i)}) -  y^{(i)})^2\Big) \\
&= \frac{1}{2m} \sum\limits^{m}_{i = 1} \Big( 2 \cdot (h_\theta(x^{(i)}) -  y^{(i)}) \cdot \frac{\partial}{\partial \theta_j}  (h_\theta(x^{(i)}) \Big)  \quad\text{(by using chain rule)}\\
&= \frac{1}{2m} \sum\limits^{m}_{i = 1} \Big( 2 \cdot (h_\theta(x^{(i)}) -  y^{(i)}) \cdot \frac{\partial}{\partial \theta_j}  (\theta_0 x_0 + \theta_1 x_1 +\dots+\theta_n x_n) \Big) \\
&= \frac{1}{m} \sum\limits^{m}_{i = 1} \Big( (h_\theta(x^{(i)}) -  y^{(i)}) \cdot \frac{\partial}{\partial \theta_j}  (\theta_0 x_0 + \theta_1 x_1 +\dots+\theta_n x_n) \Big) \\
&= \frac{1}{m} \sum\limits^{m}_{i = 1} \Big( (h_\theta(x^{(i)}) -  y^{(i)}) \cdot x_j \Big) \\
\end{align*}

If you want to read more about partial derivative, read [here](http://tutorial.math.lamar.edu/Classes/CalcIII/PartialDerivatives.aspx), if you want to know more about what effect $\alpha$ has our our gradient descent, check out ex1_multiple

Now, you can compute the gradient descent using for-loops, or you can implement them using vectorization. If you had completed all of Professor Ng's problems in MATLAB, you should have a basic understanding of [vectorization](https://www.coursera.org/learn/machine-learning/lecture/WnQWH/vectorization). Therefore, we skip straight to the vectorization implementaion of the algorithm using this formula:

\begin{equation} \theta := \theta - \frac{\alpha}{m} \big(X^{T} \cdot (X \cdot \theta - \vec{y}) \big) \end{equation}

```python
def gradient_descent(X, Y, theta, alpha, num_iters):
    m = X.shape[0]
    J_history = np.zeros((num_iters, 1))
    for x in range(0, num_iters):
        theta = theta - np.multiply((alpha / m), np.dot(X.T, np.dot(X, theta) - Y))
        J_history[x, :] = compute_cost(X, Y, theta)
    return (theta, J_history)
```

```python
iterations = 1500
alpha = 0.01
theta = gradient_descent(X, Y, theta, alpha, iterations)[0]
print('Theta value after gradient descent: ')
print(theta)
```

Now, we can visualize our linear regression and use it to predict profits:

```python
plt.figure()
plt.plot(X[:, 1], Y, 'rx', ms=10)
plt.plot(X[:, 1], np.dot(X, theta), '-')
plt.xlabel('Population of city in 10000s')
plt.ylabel('Profit in $10000')
plt.legend(['Training data', 'Linear regression'])
plt.show()

predict1 = np.dot([[1, 3.5]], theta).item()
print('For population = 35,000, we predict a profit of %.2f\n' % (predict1 * 10000))
predict2 = np.dot([[1, 7]], theta).item()
print('For population = 70,000, we predict a profit of %.2f\n' % (predict2 * 10000))
```

#### Visualizing Cost Function

To understand the purpose of our gradient descent function better, we can plot a [surface plot](https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html) and a [contour plot](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.contour.html) of our cost function relative to different values of $\theta$. 

From the two plots, we can see that the cost function is an [convex function](https://en.wikipedia.org/wiki/Convex_function), thus making it easier to determine that our minimum value is, index, a global minimum value. What gradient descent does is that it will traverse the surface to get to the lowest point of the cost function, thus minimizing it. 

```python
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-1, 4, 100)
J_vals = np.zeros((theta0_vals.size, theta1_vals.size))

for i in range(0, theta0_vals.size):
    for j in range(0, theta1_vals.size):
        t = np.array([[theta0_vals[i]], [theta1_vals[j]]])
        J_vals[i][j] = compute_cost(X, Y, t)

J_vals = J_vals.T
theta0_vals = np.reshape(theta0_vals, (theta0_vals.size, 1))
theta1_vals = np.reshape(theta1_vals, (theta1_vals.size, 1))
```

```python
# Surface plot
fig = plt.figure()
ax = fig.gca(projection='3d')
theta0_vals, theta1_vals = np.meshgrid(theta0_vals, theta1_vals)
ax.plot_surface(theta0_vals, theta1_vals, J_vals, cmap=cm.coolwarm)
plt.plot(theta[0], theta[1], 'rx', ms=10, lw=2)
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.title('Cost function surface plot')
plt.show()
```

```python
# Contour plot
plt.figure()
plt.contour(theta0_vals, theta1_vals, J_vals, np.logspace(-2, 3, 20))
plt.plot(theta[0], theta[1], 'rx', ms=10, lw=2)
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.title('Cost function contour plot')
plt.show()
```
